```python
# Huzefa Ansari | BMEN 415 | Final Project | Classification Models | SGD Classifier


# Immport Models

import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Download Files and Curate Data

dataFile = pd.read_csv ('water_potability.csv')

dataFile = dataFile.dropna()

x = dataFile.drop(columns = ['Potability']).copy()
y = dataFile ['Potability']

# Test/Train Split

x_train, x_test, y_train, y_test = train_test_split (x,y,test_size = 0.2, random_state = 42)
```

```python
# Correlation Matrix

waterpot_hm = sns.heatmap(dataFile.corr(), annot = False)
```

```python
# Import class

from sklearn.linear_model import SGDClassifier

# Fit model and Predict

SGD_Model = SGDClassifier (loss='perceptron')
SGD_Model.fit (x_train,y_train)
SGD_Prediction = SGD_Model.predict (x_test)
SGD_TrainPrediction = SGD_Model.predict (x_train)

# Metrics

from sklearn import metrics
print ("The Accuracy for the Test Model is:", metrics.accuracy_score(y_test,SGD_Prediction))
print ("The Accuracy for the Train Model is:", metrics.accuracy_score(y_train,SGD_TrainPrediction))

# Confusion Matrix

conf_mat = confusion_matrix(y_test, SGD_Prediction)
heatmap = sns.heatmap(conf_mat, annot=True, fmt='.0f')
sns.set(font_scale=1.2)
sns.set_style({'font.family':'serif', 'font.serif':'Times New Roman'})
sns.set(rc={'figure.figsize':(6,6)})
sns.set_palette("pastel")
plt.title('Correlation Matrix Heatmap')
heatmap.set_xlabel('Predicted')
heatmap.set_ylabel('True Values')
```
